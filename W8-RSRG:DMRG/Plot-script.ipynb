```python
RSvalue = np.genfromtxt("./groundstate.txt", delimiter= ",")
DMvalue = np.genfromtxt("./groundstate_DMRG_test.txt", delimiter= ",")
```

```python
#print(DMvalue)
print(RSvalue)
```

```python
def MFvalue(x):
    x = np.where( abs(x) <= 2, -1 - (x**2)/4, -abs(x))
    return  x
```

```python
def plot_eigen(data1, data2, title):

    plt.figure(figsize=(10, 5))
    plt.title(f"Energy density of the groundstate vs \u03BB")
    plt.ylabel("Groundstate (density energy)")
    plt.xlabel("\u03BB", fontsize = 15)

    plt.plot(data1[:, 0], data1[:,1], 'b', marker='.', label = f'DMRG', linestyle = '-', alpha = 0.8) #'#002FD0'
    plt.plot(data2[:, 0], data2[:,1], 'lime', marker='.', label = f'RSRG', linestyle = '-',  alpha = 0.7)
    plt.plot(list_continue, MFvalue(list_continue), 'r', marker='', label = f'MF',  alpha = 0.9)

    plt.legend()
    plt.savefig(f'./Plots/{title}.png', bbox_inches="tight")
    plt.show()
```

```python
def plot_diff(data1, data2, title):

    plt.figure(figsize=(10, 5))
    plt.title(f"Energy density of the groundstate vs \u03BB - Residue")
    plt.ylabel("Groundstate (residue w.r.t. Mean Field)")
    plt.xlabel("\u03BB", fontsize = 15)

    MF = MFvalue(data1[:, 0])
    zero = 0 * list_continue
    plt.plot(data1[:, 0], data1[:,1] - MF , 'b', marker='.', label = f'DMRG', linestyle = '-', alpha = 0.8) #'#002FD0'
    plt.plot(data2[:, 0], data2[:,1] - MF, 'lime', marker='.', label = f'RSRG', linestyle = '-',  alpha = 0.7)
    plt.plot(list_continue, zero, 'r', marker='', label = f'MF',  alpha = 0.9)

    plt.legend()
    plt.savefig(f'./Plots/{title}.png', bbox_inches="tight")
    plt.show()
```

```python
def plot_iteration(data2, title, label):

    plt.figure(figsize=(10, 5))
    plt.title(f"Number of iterations to reach convergence of order (eps) vs \u03BB")
    plt.ylabel("Number of iteration")
    plt.xlabel("\u03BB", fontsize = 15)

    plt.plot(data2[:, 0], data2[:,2] , 'b', marker='.', label = f'{label}', linestyle = '-', alpha = 0.8) #'#002FD0'
    
    plt.legend()
    plt.savefig(f'./Plots/{title}.png', bbox_inches="tight")
    plt.show()
```

```python
plot_eigen(DMvalue, RSvalue, "infiniteDMRG")
```

```python
plot_diff(DMvalue, RSvalue, "residue_GS")
```

```python
plot_iteration(RSvalue, "iterationRSRG", "RSRG, eps = 10^-15")
```

```python
plot_iteration(DMvalue, "iterationDMRG", "DMRG, eps = 10^-7")
```
